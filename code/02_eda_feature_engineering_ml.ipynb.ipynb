Tiếp theo ở trong bộ dữ liệu có những cái tên bị lỗi font chữ, vì vậy sẽ xử lý những trường hợp này.

```python
import re
import unicodedata
# Hàm chỉ chuẩn hóa phần tên, giữ nguyên phần trong ngoặc
def normalize_name_keep_suffix(name):
    match = re.match(r"^(.*?)(\s*\(.*\))$", name)
    if match:
        raw_name, suffix = match.groups()
        normalized_name = unicodedata.normalize('NFKD', raw_name).encode('ascii', 'ignore').decode('utf-8')
        return normalized_name.strip() + suffix
    else:
        # Trường hợp không có phần trong ngoặc
        return unicodedata.normalize('NFKD', name).encode('ascii', 'ignore').decode('utf-8').strip()

# Áp dụng trực tiếp lên cột PLAYER_NAME
df['PLAYER_NAME'] = df['PLAYER_NAME'].apply(normalize_name_keep_suffix)

# Hiển thị lại để kiểm tra kết quả
df['PLAYER_NAME'].head(20)
```

Tiếp theo, biến đổi các chỉ số theo tổng hết mùa sang theo trung bình từng trận, vì như vậy sẽ trực quan và rõ ràng hơn.

```python
df['MPG'] = df['MP']/df['G'] #Thời gian trung bình 1 trận

#Số lần drives
df['DRIVES/GM'] = df['DRIVES']/df['G']
df['DRIVE_FGA/GM'] = df['DRIVE_FGA']/df['G']
df['DRIVE_FG%'] = df['DRIVE_FGM']/df['DRIVE_FGA']
df['DRIVE_PASSES/GM'] = df['DRIVE_PASSES']/df['G']

#Số lần CatchandShoot
df['CATCH_SHOOT%'] = df['CATCH_SHOOT_FGM']/df['CATCH_SHOOT_FGA']
df['CATCH_SHOOT_FGA/G'] = df['CATCH_SHOOT_FGA']/df['G']
df['CATCH_SHOOT_FG3%'] = df['CATCH_SHOOT_FG3M']/df['CATCH_SHOOT_FG3A']
df['CATCH_SHOOT_FG3A/GM'] = df['CATCH_SHOOT_FG3A']/df['G']

#Số lần chạm bóng
df['TOUCHES/GM'] = df['TOUCHES']/df['G']
df['FRONT_CT_TCH/GM'] = df['FRONT_CT_TOUCHES']/df['G']
df['TIME_OF_POSS/GM (min)'] = df['TIME_OF_POSS']/df['G']

#Số cú ném pull-up
df['PULL_UP%'] = df['PULL_UP_FGM']/df['PULL_UP_FGA']
df['PULL_UP_FGA/GM'] = df['PULL_UP_FGA']/df['G']
df['PULL_UP_FG3%'] = df['PULL_UP_FG3M']/df['PULL_UP_FG3A']
df['PULL_UP_FG3A/GM'] = df['CATCH_SHOOT_FG3A']/df['G']

#Quãng đường di chuyển
df['DIST_MILES/G'] = df['DIST_MILES']/df['G']
df['DIST_MILES_OFF/GM'] = df['DIST_MILES_OFF']/df['G']
df['DIST_MILES_DEF/GM'] = df['DIST_MILES_DEF']/df['G']
```

Có thể thấy phân bố giữa các nhãn có sự chênh lệch nhưng vẫn ở mức chấp nhận được. G có số lượng lớn nhất sau đó là F và C bởi vì C không được gộp với nhãn nào còn G và F được tạo thành từ việc gộp 2 nhãn lại với nhau. Vì vậy ở đây sẽ không sử dụng các kĩ thuật cân bằng nhãn như oversampling, undersampling hay SMOTE.

Như đã đề cập ở trên, có rất nhiều cầu thủ thường chỉ đóng vai trò dự bị, vì vậy thời gian ra sân và số lượng possession rất ít, dẫn đến hiện tượng xuất hiện các giá trị NULL trong các chỉ số nâng cao hoặc phân tích theo vị trí thi đấu.
```python
#Hiệu suất ghi điểm: PTS / (FGA + FTA)
df_modelling['SCORING_RATE'] = df_modelling['PTS'] / (df_modelling['FGA'] + df_modelling['FTA'] + 1e-5)

#Hiệu quả tấn công tổng hợp
df_modelling['EFFICIENCY'] = df_modelling['TS%'] * df_modelling['USG%'] * df_modelling['ORtg']

#Mức độ hoạt động phòng ngự
df_modelling['DEF_ACTIVITY'] = df_modelling['STL'] + df_modelling['BLK'] + df_modelling['DRB']

#Hoạt động bắt-ném
df_modelling['CATCH_SHOOT_ACTIVITY'] = df_modelling['CATCH_SHOOT_FGA/G'] + df_modelling['CATCH_SHOOT_FG3A/GM']

#Mức độ dứt điểm khi kéo bóng (pull-up)
df_modelling['PULL_UP_RATIO'] = df_modelling['PULL_UP_FGA/GM'] / (df_modelling['FGA'] + 1e-5)

df_modelling['3P_COMPOSITE'] = (
    df_modelling['CATCH_SHOOT_FG3%'] * df_modelling['CATCH_SHOOT_FG3A/GM'] +
    df_modelling['PULL_UP_FG3%'] * df_modelling['PULL_UP_FG3A/GM']
 / (df_modelling['CATCH_SHOOT_FG3A/GM'] + df_modelling['PULL_UP_FG3A/GM'] + 1e-5)
)
df_modelling['POSSESSION_LOAD'] = df_modelling['TIME_OF_POSS/GM (min)'] * df_modelling['TOUCHES/GM']

df_modelling['TOTAL_REB'] = df_modelling['ORB'] + df_modelling['DRB']


```

Thay vì xử lý các giá trị outlier bằng các phương pháp như loại bỏ, nhóm sẽ log-transform để giảm sự ảnh hưởng của các giá trị ngoại lai. Vì ở NBA có một nhóm cầu thủ Superstar có chỉ số vượt trội so với các cầu thủ khác vì vậy việc bỏ đi các cầu thủ này là không hợp lý.

```python
from sklearn.preprocessing import StandardScaler
import numpy as np
import pandas as pd

# Tách X và y
X_full = df_modelling.drop(columns=['Pos'])
y = df_modelling['Pos']

# Danh sách các cột cần log-transform
log_columns = [
      'POSSESSION_LOAD',
      'VORP',
      'BLK%',
      'DRIVE_PASSES/GM',
      'DRIVE_FGA/GM',
      'DRIVES/GM',
      'ORB%',
      'PULL_UP_RATIO',
      'STL%',
      'TOV%',
      'AST%',
      'AST',
      'FTr',
      'TOV',
      'TOTAL_REB',
      'FT%'
]

# Tạo bản log-transform (X_log)
X_log = X_full.copy()
for col in log_columns:
    if col in X_log.columns:
        min_val = X_log[col].min()
        shift_val = 1e-5 if min_val >= 0 else -min_val + 1e-5
        X_log[col + '_log'] = np.log(X_log[col] + shift_val)

# Xoá cột gốc đã log
X_log.drop(columns=[col for col in log_columns if col in X_log.columns], inplace=True)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import classification_report, accuracy_score

# Dự đoán train/test
preds = {
    'KNN': {
        'train': knn.predict(X_train),
        'test': knn.predict(X_test),
    },
    'SVM': {
        'train': svm_model.predict(X_train),
        'test': svm_model.predict(X_test),
    },
    'LightGBM': {
        'train': lgbm_model.predict(X_train),
        'test': lgbm_model.predict(X_test),
    }
}

# Accuracy tổng hợp
accuracy_data = []
score_data = []

# Lấy nhãn phân lớp duy nhất
labels = sorted(set(y_train.unique()) | set(y_test.unique()))

def extract_report(report, model_name, dataset_name, labels):
    rows = []
    for label in labels:
        label = str(label)
        if label in report:
            for metric in ['precision', 'recall', 'f1-score']:
                rows.append({
                    'Model': model_name,
                    'Dataset': dataset_name,
                    'Class': label,
                    'Metric': metric,
                    'Score': report[label][metric]
                })
    return rows

# Tính toán các chỉ số
for model_name, result in preds.items():
    for dataset_name in ['train', 'test']:
        y_true = y_train if dataset_name == 'train' else y_test
        y_pred = result[dataset_name]

        # Accuracy
        acc = accuracy_score(y_true, y_pred)
        accuracy_data.append({
            'Model': model_name,
            'Dataset': dataset_name.capitalize(),
            'Accuracy': acc
        })

        # Classification report
        report = classification_report(y_true, y_pred, output_dict=True, zero_division=0)
        score_data.extend(extract_report(report, model_name, dataset_name.capitalize(), labels))

# Tạo DataFrame
df_accuracy = pd.DataFrame(accuracy_data)
df_scores = pd.DataFrame(score_data)

# Vẽ biểu đồ
fig, axes = plt.subplots(2, 4, figsize=(24, 12))

# Accuracy
sns.barplot(data=df_accuracy, x='Dataset', y='Accuracy', hue='Model', ax=axes[0, 0], palette='Set2')
axes[0, 0].set_title('Accuracy theo tập')
axes[0, 0].set_ylim(0, 1.05)
axes[0, 0].legend().set_title(None)

# Precision, Recall, F1-score (Train)
for i, metric in enumerate(['precision', 'recall', 'f1-score']):
    sns.barplot(
        data=df_scores[(df_scores['Metric'] == metric) & (df_scores['Dataset'] == 'Train')],
        x='Class', y='Score', hue='Model',
        ax=axes[0, i + 1], palette='Set2'
    )
    axes[0, i + 1].set_title(f'Train - {metric.capitalize()}')
    axes[0, i + 1].set_ylim(0, 1.05)
    axes[0, i + 1].legend().set_title(None)

# Precision, Recall, F1-score (Test)
for i, metric in enumerate(['precision', 'recall', 'f1-score']):
    sns.barplot(
        data=df_scores[(df_scores['Metric'] == metric) & (df_scores['Dataset'] == 'Test')],
        x='Class', y='Score', hue='Model',
        ax=axes[1, i + 1], palette='Set2'
    )
    axes[1, i + 1].set_title(f'Test - {metric.capitalize()}')
    axes[1, i + 1].set_ylim(0, 1.05)
    axes[1, i + 1].legend().set_title(None)

axes[1, 0].axis('off')  # Ô trống

fig.suptitle("So sánh hiệu suất KNN, SVM, LightGBM trên tập Train và Test", fontsize=18)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import classification_report, accuracy_score

# Dự đoán với các mô hình
y_pred_knn = best_knn.predict(X_test)
y_pred_svm = best_svm.predict(X_test)
y_pred_lgbm = best_lgbm.predict(X_test)

# Accuracy tổng thể
acc_knn = accuracy_score(y_test, y_pred_knn)
acc_svm = accuracy_score(y_test, y_pred_svm)
acc_lgbm = accuracy_score(y_test, y_pred_lgbm)

# Báo cáo classification
report_knn = classification_report(y_test, y_pred_knn, output_dict=True, zero_division=0)
report_svm = classification_report(y_test, y_pred_svm, output_dict=True, zero_division=0)
report_lgbm = classification_report(y_test, y_pred_lgbm, output_dict=True, zero_division=0)

# Tạo DataFrame từ classification_report
def extract_scores(report, model_name):
    data = []
    for label in y_test.unique():
        if label in report:
            for metric in ['precision', 'recall', 'f1-score']:
                data.append({
                    'Model': model_name,
                    'Class': label,
                    'Metric': metric,
                    'Score': report[label][metric]
                })
    return pd.DataFrame(data)

# Tổng hợp dữ liệu từ 3 mô hình
df_knn = extract_scores(report_knn, 'KNN')
df_svm = extract_scores(report_svm, 'SVM')
df_lgbm = extract_scores(report_lgbm, 'LightGBM')

df_scores = pd.concat([df_knn, df_svm, df_lgbm], ignore_index=True)

# Accuracy tổng hợp
df_accuracy = pd.DataFrame({
    'Model': ['KNN', 'SVM', 'LightGBM'],
    'Accuracy': [acc_knn, acc_svm, acc_lgbm]
})

# Vẽ biểu đồ
fig, axes = plt.subplots(1, 4, figsize=(22, 6))

# Accuracy
sns.barplot(data=df_accuracy, x='Model', y='Accuracy', ax=axes[0], palette='Set2')
axes[0].set_title('Accuracy tổng')
axes[0].set_ylim(0, 1.05)

# Precision, Recall, F1-score theo từng lớp
for i, metric in enumerate(['precision', 'recall', 'f1-score']):
    sns.barplot(
        data=df_scores[df_scores['Metric'] == metric],
        x='Class', y='Score', hue='Model',
        ax=axes[i + 1], palette='Set2'
    )
    axes[i + 1].set_title(metric.capitalize())
    axes[i + 1].set_ylim(0, 1.05)
    axes[i + 1].legend().set_title(None)

fig.suptitle("So sánh hiệu suất KNN, SVM, LightGBM trên từng lớp", fontsize=16)
plt.tight_layout()
plt.show()
```
